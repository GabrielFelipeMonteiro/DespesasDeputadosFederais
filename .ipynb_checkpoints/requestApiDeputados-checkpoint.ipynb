```python
import pandas as pd
import requests
from requests.exceptions import RequestException
import time
```

```python
ids = []

# url base para se conectar à API dos Dados Abertos da Câmara dos Deputados
url = "https://dadosabertos.camara.leg.br/api/v2/deputados?pagina=1&itens=100&dataInicio=2019-02-02&ordem=ASC&ordenarPor=nome"


# considerando que esta é uma API paginada, teremos que utilizar de loops para solicitarmos todos os dados necessários
while url:
    response = requests.get(url, timeout=10)
    
    if response.status_code == 200:        
        
        jsonFile = response.json()

        # adicionando à lista vazia todos os ids de deputados deta pagina da API
        ids.extend([item["id"] for item in jsonFile["dados"]])

        # verificando o conteúdo do valor retornado pela requisição, no campo de 'links'
        # se for um <'links':{'rel': 'next'}>, então nossa requisição terá que percorrer
        # para uma nova página, porém se <'rel': 'next'> não existir mais, quer dizer
        # que chegamos à pagina final. O conteúdo de <'href'> é o link para a próxima página
        next_link = next((link["href"] for link in jsonFile["links"] if link["rel"] == "next"), None)

        # sempre que existir um 'href', exemplo: "https://dadosabertos.camara.leg.br/api/v2/deputados?pagina=2&itens=100&..."
        # teremos que apontar para a próxima página da requisição.
        if next_link:
            url = next_link
            
        # quando não encontramos nova paginação, <'url'> é None, e assim o loop while é encerrado    
        else:
            url = None

    else:
        print("Conexão falhou!")

```

```python
dfDeputadosInfo = pd.DataFrame()

# Gerando valores únicos de ids de depotados
ids_unicos  = list(set(ids))

# utilizando os 'id' dos deputados como parâmetro de pesquisa à API, teremos ao final do loop,
# solicitado todas as informações básicas e cadastrais destes deputados
for id in ids_unicos:
    url = "https://dadosabertos.camara.leg.br/api/v2/deputados/{}".format(id)
    
    # A requisição de fato. Estamos batendo na porta da API.
    response = requests.get(url, timeout=10)
    
    if response.status_code == 200:        

        # Resgatando o conteúdo de response no formato json.
        jsonFile = response.json()
        
        # Pegando do json retornado, o conteúdo de 'dados'. Se não houver, será retornado um dict vazio <{}>
        dadosDeputado = jsonFile.get("dados", {})
        
        # transformando o arquivo json/dict em um DataFrame, para então alimentar o DataFrame original, exterior ao loop
        df = pd.DataFrame([dadosDeputado])
        
        # alimentando o DataFrame vazio original com o conteúdo das requisições iteradas por meio do id
        dfDeputadosInfo = pd.concat([dfDeputadosInfo, df], ignore_index=True)
        
    else:
        print("Conexão falhou!")
```

```python
# substituindo possíveis registros Nulls com um dict vazio <{}>
dfDeputadosInfo["ultimoStatus"] = dfDeputadosInfo["ultimoStatus"].apply(lambda x: {} if pd.isnull(x) else x)


# normalizando o dict presente na coluna de "ultimoStatus", expondo o conteúdo no formato estruturado (tabela)
dfUltimoStatusNormalize = pd.json_normalize(dfDeputadosInfo["ultimoStatus"])


# eliminando as tabelas que não são de nosso interesse para posteriormente
# concatenarmos este DataFrame com o DataFrame original <dfDeputadosInfo>
dfDropColumns = dfUltimoStatusNormalize.drop(columns = 
                                ["id", "uri", "nome", "data", "nomeEleitoral", "descricaoStatus", 
                                 "gabinete.nome", "gabinete.predio", "gabinete.sala", 
                                 "gabinete.andar", "gabinete.email"])
```

```python
# em um novo DataFrame auxiliar, transformamos a coluna de listas "redeSocial"
# do DataFrame original <dfDeputadosInfo> em colunas inviduais de urls de redes sociais dos deputados
# considerando que não sabemos o total de redes sociais que um deputado pode ter, utilizamos o max()
# da contagem de tamanho len() do nosso conjunto de listas para criar as colunas individuais
dfRedeSocial = pd.DataFrame(dfDeputadosInfo['redeSocial'].tolist(), 
                       columns=[f'RedeSocial{i+1}'
                       for i in range(dfDeputadosInfo['redeSocial'].apply(len).max())])
```

```python
# Gerando o DataFrame Final, com as novas colunas geradas anteriormente adicionadas ao DataFrame original
dfDeputadosInfoFinal = pd.concat([dfDeputadosInfo, dfDropColumns, dfRedeSocial],
                                 axis=1).drop(["ultimoStatus", "redeSocial"], axis=1)


```

```python
ids_unicos = list(set(ids))

dfDespesasDeputados = pd.DataFrame()

for id in ids_unicos:
    url = "https://dadosabertos.camara.leg.br/api/v2/deputados/{}/despesas?ano=2019&mes=2&itens=100&ordem=ASC&ordenarPor=ano".format(id)
    
    while url:
        try:
            reponse = requests.get(url, timeout=20)

            if reponse.status_code == 200:
                
                jsonFile = reponse.json()

                dados = jsonFile.get("dados", {})

                dfDespesas = pd.DataFrame(dados)

                dfDespesas["idDeputado"] = id

                dfDespesasDeputados = pd.concat([dfDespesasDeputados, dfDespesas], ignore_index=True)

                next_link = next((link["href"] for link in jsonFile["links"] if link["rel"] == "next"), None)

                if next_link:
                    url = next_link

                else:
                    url = None

            else:
                print("Conexão falhou!")
        except requests.exceptionsRequestException as e:
            print("Erro ao requisitar o deputado de id:{} | erro:{}".format(id, e))
            time.sleep(5)

        
```
